```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

red_wine = pd.read_csv('winequality-red.csv', sep=';')
red_wine.info()
red_wine.head()
red_wine.tail()
red_wine.describe()
red_wine.shape
```

```python
# create new column
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# transform into categorical type
red_wine['quality_label'] = pd.Categorical(red_wine['quality_label'], categories=['low', 'medium', 'high'])

red_wine
```

```python
import pandas as pd

white_wine = pd.read_csv('winequality-white.csv', sep=';')

white_wine.info()
white_wine.head()
white_wine.tail()
white_wine.describe()
white_wine.shape

```

```python
# create new column
white_wine['quality_label'] = white_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# transform into categorical type
white_wine['quality_label'] = pd.Categorical(white_wine['quality_label'], categories=['low', 'medium', 'high'])

white_wine
```

```python
# Seleccionar columnas específicas para la comparación
columns_to_compare = ['residual sugar', 'total sulfur dioxide', 'sulphates', 'alcohol', 'volatile acidity', 'quality']

# Crear resumen estadístico solo para esas columnas
red_stats = red_wine[columns_to_compare].describe()
white_stats = white_wine[columns_to_compare].describe()

# Concatenar ambos resúmenes en columnas
comparison_df = pd.concat([red_stats, white_stats], axis=1, keys=["Red Wine Statistics", "White Wine Statistics"])

# Mostrar la tabla
comparison_df
```

```python
pd.concat([red_wine.describe(), white_wine.describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])
```

```python
red_wine['wine_type'] = 'red'
white_wine['wine_type'] = 'white'

wine_df = pd.concat([red_wine, white_wine], ignore_index=True)

print(wine_df['wine_type'].value_counts())
print(wine_df.head())
```

```python
# Datos
labels = ['Red Wine', 'White Wine']
sizes = [len(red_wine), len(white_wine)] 
colors = ['#ff6666', '#ffeb99'] 

# Crear gráfico
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(sizes, colors=colors, startangle=90, wedgeprops=dict(edgecolor='white'))

plt.show()
```

```python
# Plot of white and red wine
plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")
sns.countplot(data=wine_df, x='quality', palette=['#aad7d4'])

# Título y etiquetas
plt.xlabel('Quality Score')
plt.ylabel('Number of Wines')
plt.grid(axis='y')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")

# Count plot
bars = sns.countplot(data=wine_df, x='quality', palette='Set2')

# Highlight highest
quality_counts = wine_df['quality'].value_counts().sort_index()
max_quality = quality_counts.idxmax()

for bar in bars.patches:
    if int(bar.get_x() + bar.get_width()/2) == max_quality:
        bar.set_color('#aad7d4')

plt.title('Distribution of Wine Quality Scores')
plt.xlabel('Quality Score')
plt.ylabel('Number of Wines')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Plot of RED wine
plt.figure(figsize=(8, 6))
sns.countplot(data=red_wine, x='quality')

# Título y etiquetas
plt.title('Distribution of Red Wine Quality Ratings')
plt.xlabel('Quality')
plt.ylabel('Count')

plt.show()
```

```python
# Plot of WHITE wine
plt.figure(figsize=(8, 6))
sns.countplot(data=white_wine, x='quality')

# Título y etiquetas
plt.title('Distribution of Red Wine Quality Ratings')
plt.xlabel('Quality')
plt.ylabel('Number of Wines')

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Alcohol
sns.boxplot(x='quality_label', y='alcohol', data=wine_df,
            palette='Set2', ax=axes[0])
axes[0].set_title('Wine Alcohol %')
axes[0].set_xlabel('Wine Quality Class')
axes[0].set_ylabel('Alcohol')

# pH
sns.boxplot(x='quality_label', y='pH', data=wine_df,
            palette='Set2', ax=axes[1])
axes[1].set_title('Wine pH')
axes[1].set_xlabel('Wine Quality Class')
axes[1].set_ylabel('pH')

fig.suptitle('Wine Quality - Alcohol Content / pH', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear figura y eje
fig, ax = plt.subplots(figsize=(8, 6))

# Boxplot: Alcohol vs Quality Class, separado por tipo de vino
sns.boxplot(
    data=wine_df,
    x='quality_label',
    y='alcohol',
    hue='wine_type',
    palette={'Red': '#ff6666', 'White': '#ffeb99'},
    ax=ax
)

# Títulos y etiquetas
ax.set_title('Wine Alcohol % by Quality and Type')
ax.set_xlabel('Wine Quality Class')
ax.set_ylabel('Alcohol (%)')

# Título general
fig.suptitle('Wine Quality – Alcohol Content (Red vs. White)', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Resifdual Sugar
sns.boxplot(x='quality_label', y='residual sugar', data=wine_df,
            palette='Set2', ax=axes[0])
axes[0].set_title('Wine Residual Sugar')
axes[0].set_xlabel('Wine Quality Class')
axes[0].set_ylabel('Residual Sugar')

# Volatile Acidity
sns.boxplot(x='quality_label', y='volatile acidity', data=wine_df,
            palette='Set2', ax=axes[1])
axes[1].set_title('Wine Volatile Acidity')
axes[1].set_xlabel('Wine Quality Class')
axes[1].set_ylabel('Volatile Acidity')

fig.suptitle('Wine Quality - Residual Sugar / Volatile Acidity', fontsize=14)
plt.tight_layout()
plt.show()

# No hay una relación clara entre azúcar residual y calidad del vino.
# Puede haber vinos muy dulces o secos tanto en calidades bajas como altas.
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Total Sulfur dioxide
sns.boxplot(x='quality_label', y='total sulfur dioxide', data=wine_df,
            palette='Set2', ax=axes[0])
axes[0].set_title('Wine total sulfur dioxide')
axes[0].set_xlabel('Wine Quality Class')
axes[0].set_ylabel('total sulfur dioxide')
# Relación negativa: vinos con menor dióxido de azufre suelen tener mayor calidad. 
# Esto tiene sentido, ya que un exceso de conservantes puede afectar el sabor o ser un signo de menor cuidado en la vinificación.

# Sulphates
sns.boxplot(x='quality_label', y='sulphates', data=wine_df,
            palette='Set2', ax=axes[1])
axes[1].set_title('Wine sulphates')
axes[1].set_xlabel('Wine Quality Class')
axes[1].set_ylabel('Vsulphates')

fig.suptitle('Wine Quality - total sulfur dioxide / sulphates', fontsize=14)
plt.tight_layout()
plt.show()

# Relación levemente positiva: los vinos de mayor calidad tienden a tener más sulphates, pero la diferencia no es muy marcada.
# Podría influir en el cuerpo o estructura del vino, pero no es un predictor fuerte por sí solo.
```

```python
# Calculate the correlation matrix
corr = wine_df.corr(numeric_only=True)

# Set figure size
plt.figure(figsize=(15, 8))

# Create the heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5, square = True, vmin=-1, vmax=1)

# Add title
plt.title("Correlation Heatmap of Wine Features", fontsize=16)

# Show plot
plt.show()
```

```python
# Mask upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(15, 8))
sns.heatmap(corr, mask=mask, annot=True, cmap="coolwarm", linewidths=0.5, square = True, vmin=-1, vmax=1)
plt.title("Wine Correlation Heatmap", fontsize=16)
plt.show()
```

```python
threshold = 0.3
important = corr.loc[:, (corr.abs()['quality'] > threshold)]
sns.heatmap(important, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
```

```python
# Create a new mask that matches the shape of 'important'
mask = np.triu(np.ones_like(important, dtype=bool))

selected_cols = ['alcohol', 'density', 'quality']
important = corr.loc[selected_cols, selected_cols]

sns.heatmap(important, mask=mask, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Custom Correlation Heatmap")
plt.show()
```
